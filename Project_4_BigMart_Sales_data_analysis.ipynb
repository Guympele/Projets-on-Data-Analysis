**Task** **3** : Summary statistics ( skewness and kurtosis ) for the Item_Outlet_Sales column

**Task** **4** : Let detect and remove Outliers on Item_MRP and Item_Outlet_Sales columns using Z-score.

```python
from scipy.stats import zscore

z = np.abs(zscore(df[['Item_MRP','Item_Outlet_Sales']],axis=0))
z
```

```python
np.where(z>3)
```

**Task** **5** : Let detect Outliers on Item_Outlet_Sales columns using IQR.

```python
def detect_outliers(column):
  q1 = np.quantile(df[column],0.25)
  q3 = np.quantile(df[column],0.75)
  IQR = (q3-q1)
  lower_limit = q1 - (1.5*IQR)
  upper_limit = q3 + (1.5* IQR)

  return df[(df[column]<lower_limit) | (df[column] > upper_limit)]
```

```python
detect_outliers('Item_Outlet_Sales')
```
